```python
cars_data = cars_data.replace(to_replace ="maxda", value ="mazda") 
cars_data = cars_data.replace(to_replace ="porcshce", value ="porsche") 
cars_data = cars_data.replace(to_replace ="toyouta", value ="toyota") 
cars_data = cars_data.replace(to_replace ="vokswagen", value ="volkswagen") 
cars_data = cars_data.replace(to_replace ="vw", value ="volkswagen")
cars_data = cars_data.replace(to_replace ="Nissan", value ="nissan")
```

```python
cars_data.drop(['carheight'],axis=1,inplace=True)
```

```python
cars_data.head()
```

```python
X = cars_data.drop("price", axis=1)
Y = cars_data["price"].copy()
```

```python
Y
```

```python
def binary_dummy_replace(x) :
     return x.map({"gas":1,"diesel":0,
                   "std":1,"turbo":0,
                   "two":1, "four":0,
                   "front": 1, "rear": 0})
def dummies(x,df):  
    temp = pd.get_dummies(df[x], prefix=x, drop_first = True)
    
    #l = temp.columns.values
    #for nm in l:
        #newt=x+"_"+nm
        #temp.rename({nm: Replace_Name(x)+"_"+nm}, axis=1, inplace=True)
        
    #print(temp.columns.values)
        
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
```

```python
X = dummies('symboling',X)
X = dummies('CarName',X)
X = dummies('fueltype',X)
X = dummies('aspiration',X)
X = dummies('doornumber',X)
X = dummies('carbody',X)
X = dummies('drivewheel',X)
X = dummies('enginelocation',X)

```

```python
X = dummies('enginetype',X)
X = dummies('cylindernumber',X)
X = dummies('fuelsystem',X)
```

```python
X
```

```python
cars_data
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    #     ..... add as many as you want in your pipeline
    ('std_scaler', StandardScaler()),
])
```

```python
X = my_pipeline.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X_train
```

model selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model1 = LinearRegression()
model2 = DecisionTreeRegressor()
model3 = RandomForestRegressor()
model1.fit(X_train, Y_train)
model2.fit(X_train, Y_train)
model3.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(some_data_y, predict_y)
rmse = np.sqrt(mse)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model1, X_train, Y_train, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```
